```python
import pandas as pd
from datetime import datetime
import soccerdata as sd
import os,sys,requests
import sql_connector as SQL
import config
```

```python
matches = SQL.execute_select_query("select * from match_schedule")
```

```python
matches = pd.DataFrame(matches)
```

```python
matches.head(2)
```

```python
def check_WSurl_contains_id(row):
    if str(int(row['whoscored_id'])) in str(row['whoscored_url']):
        return True
    else:
        return False
def check_FBurl_contains_id(row):
    if str(row['fbref_id']) in str(row['fbref_url']):
        return True
    else:
        return False
                
def check_FMurl_contains_id(row):
    if str(row['fotmob_id']) in str(row['fotmob_id']):
        return True
    else:
        return False
```

```python
df = pd.DataFrame()
# Apply the function to each row and create a new column 'contains_id'
df= matches[['id','competition','season','stage','date','home_team','away_team','score','whoscored_id','whoscored_url']]
df['WS_contains_id'] = matches.apply(check_WSurl_contains_id, axis=1)
df['FB_contains_id'] = matches.apply(check_FBurl_contains_id, axis=1)
df['FM_contains_id'] = matches.apply(check_FMurl_contains_id, axis=1)
df['whoscored_id'] = df['whoscored_id'].astype(int)
# Display the DataFrame with the new column
df
```

```python
df['WS_contains_id'].value_counts(),\
df['FB_contains_id'].value_counts(),\
df['FM_contains_id'].value_counts()
```

```python
def extract_id_from_url(row):
    # Define the regex pattern to match the ID in the URL
    pattern = r'/Matches/(\d+)/'
    import re
    # Search for the pattern in the URL
    match = re.search(pattern, row['whoscored_url'])

    if match:
        # Extract the ID from the matched group
        return match.group(1)
    else:
        # If no match found, return None
        return None
```

```python
df['correct_ws_id'] = df.apply(extract_id_from_url,axis=1)
```

```python
for ws_id,id in zip(df['correct_ws_id'],df['id']):
    # print(ws_id,id)
    # print("select * from match_schedule where id = '{}'".format(id))
    r = SQL.execute_select_query("select * from match_schedule where id = '{}'".format(id))
    if len(r) == 1:
        r = r[0]
        q = "update match_schedule set whoscored_id = '{}' where id = '{}'".format(ws_id,id)
        # print(q)
        SQL.execute_update_query(q)
        print(f"Completed for id : {r['id']}")
    # break
```
